```python
import pandas as pd
from keras.datasets import cifar100
```

```python
cifar100.load_data()
```

```python
(train_i, train_l), (test_i, test_l) = cifar100.load_data(label_mode='fine')
df_train_i=pd.DataFrame(train_i.reshape(train_i.shape[0],-1))
df_train_i.head()
```

```python
train_i.shape
```

```python
train_l.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import models, layers
network=models.Sequential()
network.add(layers.Dense(512,activation='relu',input_shape=(32*32*3,)))
network.add(layers.Dense(100,activation='softmax'))
```

```python
from keras.optimizers import RMSprop  # Import the RMSprop optimizer

network.compile(optimizer=RMSprop(learning_rate=0.0001),
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
from keras.utils import to_categorical
train_l=to_categorical(train_l)
test_l=to_categorical(test_l)
```

```python
train_i=train_i.reshape(50000,32*32*3)
train_i=train_i.astype('float32')/255
test_i=test_i.reshape(10000,32*32*3)
test_i=test_i.astype('float32')/255
```

```python
network.fit(train_i,train_l,batch_size=200,epochs=5)
```

```python
test_loss,test_accuracy=network.evaluate(test_i,test_l)
```
