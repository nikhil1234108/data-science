```python
labels=['python','c','c++','java']
numbers=[75,85,95,65]
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax1.pie(numbers,labels=labels,autopct='%.2f%%',colors=['r', 'g', 'b', 'c'])
ax1.legend()
ax1.set_title("pie chart")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
plt.show()
```

```python
labels=["circket","football","tennis","basketball"]
score=[99.98,88,77,77]
Explode=(0.1,0,0,0.2)
fig=plt.figure(figsize=(10,10))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.pie(score, labels=labels, autopct='%.2f%%', colors=['r', 'g', 'b', 'c'],explode=Explode)
ax1.legend(loc=10, bbox_to_anchor=(1, 1))
ax1.set_title("pie chart")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(y="total_bill",data=df)
plt.show()
```

```python
q1=df['total_bill'].quantile(0.25)
q2=df["total_bill"].quantile(0.5)
q3=df["total_bill"].quantile(0.75)
iqr=q3-q1
upper_bound=q3+1.5*iqr
lower_bound=q1-1.5*iqr
print(upper_bound)
print(lower_bound)
```

```python
df_filtered = df[(df['total_bill'] >= lower_bound) & (df['total_bill'] <= upper_bound)]

# Display the new DataFrame without outliers
print("\nDataFrame after removing outliers:")
print(df_filtered)

# Optionally, display a boxplot of the cleaned data
sns.boxplot(x="total_bill", data=df_filtered)
plt.show()
```
