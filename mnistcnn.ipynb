```python
import pandas as pd
from keras.datasets import mnist
(train_i,train_l),(test_i,test_l)=mnist.load_data()
df_train_i=pd.DataFrame(train_i.reshape(train_i.shape[0],-1))
df_train_i
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import models, layers
import tensorflow as tf
network=models.Sequential()
network.add(layers.Dense(512,activation='relu',input_shape=(28*28,)))
network.add(tf.keras.layers.Flatten())
network.add(layers.Dense(10,activation='softmax',))
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
train_i = train_i.reshape(60000, 28*28)
train_i = train_i.astype('float32')/255
```

```python
test_i = test_i.reshape(10000, 28*28)
test_i = test_i.astype('float32')/255
```

```python
from keras.utils import to_categorical
train_l = to_categorical(train_l)
test_l = to_categorical(test_l)
```

```python
train_i
```

```python
train_l
```

```python
network.fit(train_i, train_l, batch_size = 128, epochs=2)
```

```python
test_loss, test_accuracy = network.evaluate(test_i, test_l)
```

```python
print("loss",test_loss)
print("accuracy",test_accuracy)
```
