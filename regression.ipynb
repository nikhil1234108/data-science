```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 2 * x + 3

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Data')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = x**2

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-Linear Data')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([[1500], [2000], [2500], [3000], [3500]])
y = np.array([300000, 400000, 500000, 600000, 700000])

# Linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict
X_test = np.array([[2250]])
y_pred = model.predict(X_test)
print(f"Predicted price for 2250 sq ft: ${y_pred[0]}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data (days and number of infections)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([10, 50, 200, 800, 3000])

# Polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict
X_test = np.array([[6]])
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)
print(f"Predicted infections on day 6: {y_pred[0]}")
```
